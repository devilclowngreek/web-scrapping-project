## Web Scrapping Kost Jakarta Project ##

```python
from bs4 import BeautifulSoup
import requests, time, datetime

#connect to website
url = 'https://www.kostjakarta.com/category/kost-jakarta-pusat/'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}

page = requests.get(url, headers=headers)

soup1 = BeautifulSoup(page.content,"html.parser")
soup2 = BeautifulSoup(soup1.prettify(), "html.parser")

data = []
for i in soup2.find('li', class_=False, id=False):
    title = soup2.findAll('li', class_=False, id=False)
    data.append(title)
print(data)
# title = soup2.find(class_='entry-content')
# data = []
# for i in title:
#     title2 = soup2.find('li', class_=False, id=False)
#     data.append(title2)

# print(data)
```
